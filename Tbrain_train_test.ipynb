## submission.csv之轉檔範例

* dat_cust_no 是只有顧客名稱的dataframe，我把他後面直接併cc_test, fx_test, ln_test, wm_test 的預測結果，變成最後的檔案 (注意 我這邊已經確認過預測結果的順序跟cust_no的順序相同，因此才可以直接concat)

`submission = pd.concat([dat_cust_no,pd.DataFrame({'CC_IND':cc_test,
                                                  'FX_IND':fx_test,
                                                  'LN_IND':ln_test,
                                                  'WM_IND':wm_test,})], axis =1)`

* 存成名稱叫做'submission0124_1.csv'的檔案 (有 'CUST_NO','CC_IND','FX_IND','LN_IND','WM_IND' 五個欄位)

`submission.to_csv('submission0124_1.csv', index = False)`

## 資料集說明 
* Link: https://tbrain.trendmicro.com.tw/Competitions/Details/5
* **訓練資料集時間區間**為 9447+1 (**9448**) ~ 9447+120 (**9567**) 之間，共120日之間產品交易/申請資訊．
* 本次比賽是請參賽者**預測時間區間**為在9447+121 (**9568**) ~ 9447+150 (**9597**) 時間區間內（共30日），顧客在四個產品線CC_IND, FX_IND, LN_IND, WM_IND是否會進行交易，**若預測顧客該時間內於任一產品線交易一次以上，該產品線的預測值為1，否則為0．** 例如: 顧客序號12345，預測信用卡核卡1次，外匯交易2次，沒有信貸申請，信託交易3次，該筆預測資料應為: 12345,1,1,0,1 ．
* TBN_Y_ZERO.csv則是提供給參賽者做為上傳範例資料檔使用，內部預設值是全部為0．
